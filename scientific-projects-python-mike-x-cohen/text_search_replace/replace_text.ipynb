## Text search and replace

```python
import string
import numpy as np
```

```python
with open('data/captions_text.vtt') as caption_text:
    text = caption_text.read()

text = text[7:]
```

```python
str_chars = string.ascii_letters
str_chars = str_chars
str_chars = [char for char in str_chars]
words = []

for i, letter in enumerate(text.split()):
    if letter[0] in str_chars:
        words.append(letter)

for i in range(len(words)):
    if len(words[i]) == 4 and not words[i][-1] == '.':
        words[i] = words[i][0] + '***'

# join the words into string
clean_text = ' '.join(words)
```

```python
with open('data/clean_captions.txt', 'w') as file:
    file.write(clean_text)
```

### Bonus Exercise

Randomize every character inside word, other than first and last character.

```python
text = "I have often wondered whether its possible to read an entire book faster if the useless letters were removed."

words_list = text.split()


for i in range(len(words_list)):
    if len(words_list[i]) >= 4:
        new_random_word_list = list(range(len(words_list[i])))
        # not include the first char and last char
        random_permutations = np.random.permutation(len(words_list[i]) - 1)

        char_index = 1
        permutation_index = 0
        while permutation_index < len(words_list[i]) - 1:
            if random_permutations[permutation_index] != 0:
                new_random_word_list[char_index] = words_list[i][random_permutations[permutation_index]]
            else:
                char_index -= 1
            char_index += 1
            permutation_index += 1

        # add the first and last letter
        new_random_word_list[0] = words_list[i][0]
        # new_random_word_list[len(new_random_word_list) - 1] = words_list[i][len(words_list[i]) - 1]
        new_random_word_list[-1] = words_list[i][-1]

        words_list[i] = ''.join(new_random_word_list)


' '.join(words_list)
```

#### Second Solution

```python
words = text.split()

for wi in range(len(words)):
    if len(words[wi]) < 4:
        continue

    # from first char to last
    this_word = words[wi][1:-1]
    index = np.random.permutation(len(this_word))
    words[wi] = words[wi][0] + ''.join([this_word[i] for i in index]) + words[wi][-1]

    ans_str = ' '.join(words)

ans_str
```

#### Replace with random letters

```python
all_letters = string.ascii_lowercase
words = text.split()

for wi in range(len(words)):
    if len(words[wi]) < 4:
        continue

    # from first char to last
    this_word = words[wi][1:-1]
    index = np.random.randint(0, len(all_letters), len(this_word))
    words[wi] = words[wi][0] + ''.join([all_letters[i] for i in index]) + words[wi][-1]

    ans_str = ' '.join(words)

# this is not readable
ans_str
```
