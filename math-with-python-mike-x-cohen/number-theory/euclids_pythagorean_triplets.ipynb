## Pythagorean theorem
$$c^2 = a^2 + b^2$$

## Pythagorean Triplets
Sets of integers that works in pythagorean theorem.

### Examples:
(5,3,4) are pythagorean triplets
$$5^2 = 3^2 + 4^2 \quad \Rightarrow \quad 25 = 9 + 16$$

(13,5,12) are pythagorean triplets
$$13^2 = 5^2 + 12^2 \quad \Rightarrow \quad 169 = 25 + 144$$


## Euclid's Formula to find triplets

$a = m^2 - n^2$

$b = 2mn$

$c = m^2 + n^2$

### Constraints

"m > n"

"m,n" are between 2, and 50 (we have add this constraint ourselves for our own purposes)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numberrange = np.arange(2,51)
a = []
b = []
c = []

for m in numberrange:
    for n in numberrange:
        # if we ignore this constraints, we can get the negative values, SEE the result in every notebook, by removing this constraint
        if m > n:
            a.append(m**2 - n**2)
            b.append(2*m*n)
            c.append(m**2 + n**2)
```

```python
# plotting 
plt.scatter(a,b,c=c,marker='o',alpha=.6,linewidths=.5,edgecolors="k")

plt.show()
```

```python
# without constraints
numberrange = np.arange(2,51)
a = []
b = []
c = []

for m in numberrange:
    for n in numberrange:  
        a.append(m**2 - n**2)
        b.append(2*m*n)
        c.append(m**2 + n**2)

# plotting 
plt.scatter(a,b,c=c,marker='o',alpha=.6,linewidths=.5,edgecolors="k")

plt.show()
```

#### Check if the calculated a,b,c are really fulfill the pythagorean theorem

```python
pythagoreancheck = np.array(a)**2 + np.array(b)**2 - np.array(c)**2

plt.plot(pythagoreancheck, "o")
plt.show()
```
