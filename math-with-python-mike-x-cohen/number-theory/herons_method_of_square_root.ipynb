## Heron's Method

It is one of the simple methods of finding the square root of a number.

$$\sqrt{Y} = X \quad \Rightarrow \quad Y = X^2$$

"x" is closer to "X", and "$\epsilon$" is a very small value.

$$Y = (x + \epsilon)^2$$

Solve for "$\epsilon$".

$$Y = (x + \epsilon)(x + \epsilon)$$

$$Y = x^2 + \epsilon^2 + 2x\epsilon$$

$$Y - x^2 = \epsilon^2 + 2x\epsilon$$

$$Y - x^2 = (\epsilon + 2x)\epsilon$$

Because "$\epsilon$" in the brackets is very small we just ignore it.

$$\frac{Y - x^2}{2x} \approx \epsilon$$

Add value of "$\epsilon$" in the second equation, and then add value of Y in first equation, then "X" becomes.

$$X \approx x + \frac{Y - x^2}{2x}$$

"x" is approximately equal to "X".

$$x \leftarrow x + \frac{Y - x^2}{2x}$$

Take LCM, and solve further.

$$x \leftarrow \frac{x + Y/x}{2}$$

$\text{Because } x \approx X = \sqrt{Y}$, after each iteration, "x" becomes better and better, hence becomes approximately equal to "X".

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Y = 100
x = Y/3

# after each iteration, x will become closer, and closer to the original value of square root of "Y"
for n in range(5):
    x = (x + Y/x) / 2

x
```

```python
nums2sqrt = np.linspace(2,101,50)
niterations = np.arange(3,9)

err = np.zeros((len(niterations),len(nums2sqrt)))

# loop over numbers to square root
for ni,num in enumerate(nums2sqrt):
    Y = num
    # loop over the number of iterations
    for ii,iters in enumerate(niterations):
        # Herons' Algorithm
        x = Y/3
        for n in range(iters):
            x = (x + Y/x) / 2

        # get the error
        err[ii,ni] = abs(x - np.sqrt(Y))
```

```python
plt.imshow(-np.log(err),aspect=10,extent=[nums2sqrt[0],nums2sqrt[-1],niterations[-1],niterations[0]])

plt.show()
```

```python
num2sqr = 13
start_est = np.linspace(-num2sqr,num2sqr,40)

niters = 5
# initialize a result matrix
sqrt_alg_results = np.zeros((len(start_est),niters+1))

# loop over starting estimates
for idx,startnum in enumerate(start_est):
    # initialize the starting value
    x = [startnum]

    # Herons' Algorithm
    for n in range(niters):
        new_x = (x[n] + num2sqr/x[n]) / 2
        x.append(new_x)

    sqrt_alg_results[idx,:] = x


fig,ax = plt.subplots(1,figsize=(7,5))
plt.plot(np.arange(0,niters+1),sqrt_alg_results.T)
plt.show()
```
