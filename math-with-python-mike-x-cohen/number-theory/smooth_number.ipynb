## k-smooth number

A positive integer whose largest prime factor is $\le$ k.

Prime Factors of 30 are 2,3,5

$$30 = 2 \times 3 \times 5$$

The largest prime-factor is 5, so 30 is "5-smooth", but 30 is also "7-smooth", "11-smooth", "13-smooth".

Because if we set 7,11,13,... (prime numbers) as "k", so the largest prime factor of 30 (5), will also be less than these numbers (7,11,13,...), so 30 is also these_number-smooth.

And 30 is not a 3-smooth number, because, if we set 3 as "k", largest prime factor of 30 (5) will not be less than "k" (3)

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
max_n = 10000
largest_prime_fact = np.zeros(max_n+1,dtype=int)/0

for i in range(2,max_n + 1):
    largest_prime_fact[i] = np.max(sym.primefactors(i))


plt.plot(largest_prime_fact,'k.',markersize=.5)
plt.xlabel("Number")
plt.xlabel("Largest Prime factor")

plt.show()
```

```python
# count number of occurrences of max prime factors
unique_max_primes = np.unique(largest_prime_fact)
counts = np.zeros(len(unique_max_primes),dtype=int)

for i,u in enumerate(unique_max_primes):
    counts[i] = np.sum(largest_prime_fact==u)


plt.plot(unique_max_primes,counts,'k.')

plt.xlim([-2,500])
plt.title("%s is the modal max prime factor" %unique_max_primes[np.argmax(counts)])
plt.show()
```

## Exercise

```python
numberlist = np.arange(max_n+1)
smooth_count = np.zeros(len(unique_max_primes))

for i,u in enumerate(unique_max_primes):
    smooth_count[i] = len(numberlist[largest_prime_fact<=u])

plt.plot(unique_max_primes,smooth_count,'k.')

plt.xlim([-2,500])
plt.xlabel("Smooth Number")
plt.ylabel("Number of Numbers")
plt.title("k-smooth number frequencies")
plt.show()
```
