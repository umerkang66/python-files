## Fermat's Theorem
This theorem states that 

$$x^n + y^n = z^n$$

Has no integer solution for "n>2", when "n=2", it is actually pythagorean theorem. If "n" is greater than 2, then it is impossible to separate, a cube into two cubes, or a fourth power into two fourth powers, or in general, any power higher than the second. Thus there will not an integer solution. It can be floating point number or other.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
inits = np.arange(1,101)
Z = np.zeros((len(inits),len(inits)))
for x in inits:
    for y in inits:
        Z[x-1,y-1] = (x**3 + y**3)**(1/3)

z_inits = Z%1==0
```

```python
# visualize
fig,ax = plt.subplots(1,2,figsize=(12,6))

ax[0].imshow(Z)
ax[0].set_xlabel("y")
ax[0].set_ylabel("x")
ax[0].set_title("$z = \\sqrt[3]{x^3+y^3}$",fontsize=18)

ax[1].imshow(z_inits)
ax[1].set_xlabel("y")
ax[1].set_ylabel("x")
ax[1].set_title("Whether '$z$' is an int", fontsize=18)

plt.show()
```
