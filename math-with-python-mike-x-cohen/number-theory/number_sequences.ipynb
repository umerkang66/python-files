```python
import numpy as np
import matplotlib.pyplot as plt
```

## Convergent Sequences
A sequence is a convergent if it points to (or converges into) the single number, as the x-axis get larger and larger.

## Divergent Sequences
A sequence is a divergent if it goes away from single number (or diverges from) the single number, as the x-axis get larger and larger.

```python
# convergent sequences
s1 = np.array([])
s2 = np.array([])
s3 = np.array([])
s4 = np.array([])

xx = np.arange(1,81)

for a in xx:
    s1 = np.append(s1,a/(a+1))
    s2 = np.append(s2,(a+1)/a)
    s3 = np.append(s3,a/sum(np.arange(1,a+1)))
    s4 = np.append(s4,(-1)**a/np.sqrt(a))

fig,ax = plt.subplots(1,figsize=(8,6))
plt.plot(xx,s1,label="$a/(a+1)$")
plt.plot(xx,s2,label="$(a+1)/a$")
plt.plot(xx,s3,label="$a/\Sigma a$")
plt.plot(xx,s4,label="$-1^2/\\sqrt{a}$")

plt.legend(loc="upper right",fontsize=16)
plt.show()
```

```python
# divergent sequences
s1 = np.array([])
s2 = np.array([])
xx = np.arange(1,81)

for a in xx:
    s1 = np.append(s1,sum(np.arange(1,a+1))/a)
    s2 = np.append(s2,np.sqrt(a)/(-1)**a)

fig,ax = plt.subplots(1,figsize=(8,6))
plt.plot(xx,s1,label="$\Sigma a/a$")
plt.plot(xx,s2,label="$\\sqrt{a}/-1^2$")

plt.legend(loc="upper right",fontsize=16)
plt.show()
```
