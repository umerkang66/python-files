## Euler's Formula

$\mathrm{e}^{ik} = \cos(k) + i\sin(k)$

where "k" is the angle, "i" is the imaginary part, "e" is the natural exponent

To create the line outside of the unit circle, we use this formula

$\mathrm{e}^{ik} = m (\cos(k) + i\sin(k))$

where "m" is the distance from (0,0) coordinate to where that line should be plotted


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# pi/4 ==> 45 deg
k = np.pi / 6
m = 2.3

# left part of euler formula
eul = m*np.exp(1j*k)
# right part of euler formula
cis = m * (np.cos(k) + 1j * np.sin(k))

# these should be identical
print(eul)
print(cis)
```

#### Plotting graph

```python
mag = np.abs(eul) # == m
ang = np.angle(eul) # == k

plt.polar([0,ang],[0,mag],'r')
plt.polar(k,m,'bo')

plt.show()
```
