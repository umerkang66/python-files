```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x = np.linspace(0,6*np.pi,400)

plt.plot(x,np.sin(x),'r', label="$\\sin(\\theta)$")
plt.plot(x,np.cos(x),'b', label="$\\cos(\\theta)$")

plt.legend()
plt.xlabel("rad.")
plt.ylabel("function value")

plt.show()
```

$\tan(\theta) = \frac{\sin(\theta)}{\cos(\theta)}$

```python
theta = np.linspace(0,4*np.pi,300)

plt.plot(theta,np.tan(theta), 'k')
plt.ylim([-5,5])
plt.show()
```

```python
# to get the random numbers in radian, we have to multiple it by 2pi
ang = np.random.rand()*2*np.pi

tan = np.tan(ang)
sine_consine = np.sin(ang) / np.cos(ang)

print(ang,tan-sine_consine)
```

$\cos(\theta)^2 + \sin(\theta)^2 = 1$

```python
thetas = np.linspace(0,2*np.pi,13)

sin = np.sin(thetas)
cos = np.cos(thetas)
ans = sin**2 + cos**2

print(ans)
```

y1 = $\sin(\theta + \cos(\theta))$

y2 = $\cos(\theta + \sin(\theta))$

```python
angles = np.linspace(0, 2*np.pi, 105)

y1 = np.sin(angles + np.cos(angles))
y2 = np.cos(angles + np.sin(angles))

plt.plot(angles,y1)
plt.plot(angles,y2)

plt.legend(["$\\sin(\\theta + \\cos(\\theta))$", "$\\cos(\\theta + \\sin(\\theta))$"])
plt.show()
```

```python
theta_sym = sym.pi*7/6
theta = np.pi*7/6 # theta = float(theta_sym)

# plot the unit circle
x = np.linspace(0,2*np.pi,200)
plt.plot(np.cos(x),np.sin(x), color='gray')
# dash lines at the coordinates
plt.plot([0,0],[-1.1,1.1], '--', color="gray")
plt.plot([-1.1,1.1],[0,0], '--', color="gray")

# arc: go from 0 to where the angle stops
y = np.linspace(0,theta,200)
plt.plot(np.cos(y),np.sin(y), 'k', linewidth=3)
# vector goes from 0 to angle in straight line
plt.plot([0,np.cos(theta)],[0,np.sin(theta)], 'k:')
# black circle
plt.plot(np.cos(theta),np.sin(theta), 'ko')

# plot the separate sine/cosine components
plt.plot([0,0],[0,np.sin(theta)], 'r', label="sin part")
plt.plot([np.cos(theta),0],[np.sin(theta),np.sin(theta)], 'r:')

plt.plot([0,np.cos(theta)],[0,0], 'g', label="cos part")
plt.plot([np.cos(theta),np.cos(theta)],[0,np.sin(theta)], 'g:')

plt.title("$\\theta = %s$" %sym.latex(theta_sym))
plt.legend()
plt.axis("square")
plt.show()
```
