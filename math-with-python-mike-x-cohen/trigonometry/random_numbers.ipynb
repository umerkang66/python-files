```python
import numpy as np
import matplotlib.pyplot as plt
```

## Uniformly distributed random numbers

```python
nums = np.random.rand(1000)
min_val = 2
max_val = 17
nums = nums * (max_val - min_val) + min_val

plt.plot(nums, 'gs')
plt.show()

plt.hist(nums)
plt.show()
```

## Normally distributed random numbers

There will be lot of random numbers (in quantity) close to 0, as we go upwards, or downwards (positive or negative numbers) the random numbers (in quantity) starts to decrease in number

```python
nums = np.random.randn(1000)

plt.plot(nums, 'gs',alpha=.5)
plt.show()

plt.hist(nums)
plt.show()
```

## Exercise

#### Exercise_1

```python
desired_mean = 15
desired_standard_deviation = 4.3

nums = np.random.randn(1000) 
# change the default mean to exactly close to zero
nums = nums - np.mean(nums)
# change the default std deviation exactly close to zero
nums = nums / np.std(nums)

# change the standard deviation by multiplying is
# change the mean by adding it
nums = nums * desired_standard_deviation + desired_mean

print("This distribution has a mean of %s and a standard deviation of %s" %(np.mean(nums), np.std(nums)))

plt.subplot(121)
plt.plot(nums, 'gs',alpha=.5)

plt.subplot(122)
plt.hist(nums)

plt.show()
```

```python
angles_n = np.random.randn(100)

for i in angles_n: 
    plt.plot([0,np.cos(i)],[0,np.sin(i)])

plt.axis([-1,1,-1,1])
plt.axis("square")
plt.show()

angles_i = np.random.rand(100)*2*np.pi

for i in angles_i: 
    plt.plot([0,np.cos(i)],[0,np.sin(i)])

plt.axis([-1,1,-1,1])
plt.axis("square")
plt.show()
```
