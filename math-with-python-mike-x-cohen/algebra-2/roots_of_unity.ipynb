```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

## Roots of Unity

How many numbers (real or complex) will give 1, when their power will be "n" (given number)

The formula to find the roots of unity is

$z = e^{\frac{2\pi ik}{n}}$

where "n" is the number that will be in power position, "k" is numbers starting from 0 to n - 1, and "i" is the 1j (1 complex number)

```python
n = 4

for k in range(0, n): 
    z = sym.exp( 2*sym.pi*sym.I*k/n)
    display(Math("(%s)^{%s} \\Rightarrow %s" %(sym.latex(z), n, sym.latex(z**n))))
```

#### Plotting the roots of unity

```python
n = 14

for k in range(0, n): 
    z = np.exp( 2*np.pi*1j*k/n)
    plt.plot([0,np.real(z)],[0,np.imag(z)])

plt.axis("square")
x = np.linspace(0, 2*np.pi, 100)
plt.plot(np.cos(x), np.sin(x), color="gray")

plt.show()
```
