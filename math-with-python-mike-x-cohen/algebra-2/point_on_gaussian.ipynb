## Gaussian Curve

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Gaussian Formula is 

$y=e^\frac{-4\log(2)t^2}{h^2}$

where "e" is the natural exponent, "t" is vector, a list of time points

we draw a line at 50% of the gaussian peak horizontally, where it dissects the gaussian curve, we draw two lines to the bottom (x-axis), the distance between these two lines is "h"

```python
h = .999
t = np.linspace(-3,3,1001)

# gaussian formula
g = np.exp((-4*np.log(2)*t**2) / (h**2))

# plot
plt.plot(t,g)

plt.xlabel('t')
plt.ylabel('g')
plt.title("Gaussian with FWHM=" + str(h))

plt.show()
```

```python
y_val_2_find = .5
t_val = h*np.sqrt(np.log(y_val_2_find)/(-4*np.log(2)))

print(t_val)
```

## Exercise

```python
N = 100

matrix = np.zeros((N,len(t)))
h = np.zeros(N)

for i in range(N): 
    h[i] = (i+1)/N
    matrix[i,:] = np.exp((-4*np.log(2)*t**2) / (h[i]**2))

plt.pcolormesh(t,h,matrix,shading='auto')

plt.xlabel('t')
plt.xlabel('h')

plt.show()
```
