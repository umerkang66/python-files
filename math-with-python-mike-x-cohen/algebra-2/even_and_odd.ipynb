```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Even Function

A function is even function if the graph (function answer) is the same, no matter if function values are positive or negative

```python
x = np.linspace(-5,5,20)
f_even = x**2
f_even_neg = (-x)**2

plt.plot(x, f_even)
plt.plot(x, f_even_neg, 'ro')

plt.show()
```

#### Odd function

A function is odd function if the graph (function answer) is not the same, if function values are positive and negative

And if function of negative values and negative function of positive values is same, then also the function is odd

```python
f_odd = x**3
f_odd_neg = (-x)**3
# entire function negative
f_neg_odd = -f_odd

plt.plot(x,f_odd)
plt.plot(x,f_odd_neg, 'ro')
plt.plot(x,f_neg_odd, 'g')

plt.show()
```

## Exercise

Check whether the sine, and cosine are even or odd functions

```python
# cosine
theta = np.linspace(-2*np.pi, 2*np.pi, 39)

cos_fun = np.cos(theta)
cos_fun_neg = np.cos(-theta)

plt.plot(theta,cos_fun)
plt.plot(theta,cos_fun_neg, "ro")
plt.title("$\\cos(\\theta): even$")

plt.show()
```

```python
# sine
theta = np.linspace(-2*np.pi, 2*np.pi, 39)

sin_fun = np.sin(theta)
sin_fun_neg = np.sin(-theta)

plt.plot(theta,sin_fun)
plt.plot(theta,sin_fun_neg, "ro")
plt.title("$\\sin(\\theta): odd$")

plt.show()
```
