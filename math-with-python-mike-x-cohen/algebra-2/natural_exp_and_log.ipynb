```python
import numpy as np
import matplotlib.pyplot as plt
```

## Natural Exponents

```python
x = np.linspace(-3,3,99)

plt.plot(x, np.exp(x))
plt.xlabel("x")
plt.ylabel("$e^x$")
plt.show()

plt.plot(x, np.log(x))
plt.xlabel("x")
plt.ylabel("$log(x)$")
plt.show()
```

## Exercise

```python
x = np.linspace(.0001, 10, 10)

# y1 will be x, cause log, and exp cancel each other
y1 = np.log(np.exp(x))
# y2 will be x, cause exp, and log cancel each other
y2 = np.exp(np.log(x))

# it is just the plotting of x,x
plt.plot(x,y1,"b", label="$\\log(e^x)$")
plt.plot(x,y2,"yo", label="$e^{\\log(x)}$")

plt.axis("square")
plt.legend()

plt.show()
```
