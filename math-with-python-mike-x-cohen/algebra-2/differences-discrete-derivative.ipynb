```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 4]

# subtract every element from its previous one, and return numpy array
np.diff(x)
```

```python
y = np.arange(0, 10)

print(np.diff(y))

# "2" means take the difference two times, one for the "y" array, and second for returned array from the previous one
print(np.diff(y, 2)) # same as np.diff(np.diff(y,2))
```

```python
x = np.linspace(-2, 2, 101)
f = x**2
df = np.diff(f)
dx = np.diff(x)

plt.plot(x, f, label="f")
plt.plot(x[:-1], df/dx, label="df/dx")

plt.axis([-2, 2, -1, 2])
plt.grid()
plt.legend()
plt.show()
```
