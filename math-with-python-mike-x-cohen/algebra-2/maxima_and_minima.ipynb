#### Maximum example

```python
x = np.linspace(0, 2*np.pi, 20)
fx = -(np.cos(x) + np.sqrt(x))

# max value in f_max array
f_max = np.max(fx)
# max index in f_max array
f_max_index = np.argmax(fx)

print(f_max, f_max_index)

plt.plot(x,fx, "bo-")
plt.plot(x[f_max_index], f_max, "rs")
plt.show()
```

#### Minimum Example

```python
x = np.linspace(-2, 2, 51)
fx = x**3 + x**4

f_min_index = np.argmin(fx)

plt.plot(x,fx,"bs-")
plt.plot(x[f_min_index],fx[f_min_index], "rs")

# cut of the graph
plt.xlim([-1.2,.5])
plt.ylim([-.2,.2])

plt.show()
```

```python
from scipy.signal import find_peaks
```

```python
x = np.linspace(0, 12*np.pi,213)
fx = -np.cos(x) - np.sqrt(x)

# this will return index of maximum elements
peaks = find_peaks(fx)[0]

plt.plot(x,fx)
plt.plot(x[peaks],fx[peaks], 'o')

plt.show()
```
