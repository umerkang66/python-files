##  Definite Integral

Definite Integral can be find by multiplying the width and the height of small portions with respect to x-axis in graph of the given function

$\int^b_a f(x) dx$

$f(x)$ is height at each point, and dx is the width of every small portion from a, to b, on the graph of the given function

## Indefinite Integral

Integral is the opposite of the derivative, whatever the output of integral of expression is, if we take the derivative of output, we get the original expression

$\int bx^a dx = \frac{bx^{a+1}}{a+1} + c$

if we take the derivative of expression $\frac{bx^{a+1}}{a+1} + c$, we will get the answer "$bx^a$", derivative of constant is 0, so c = 0

$\int 4x^2 dx = \frac{4x^3}{3} + c$

if we take the derivative of expression $\frac{4x^3}{3} + c$, we will get the answer "$4x^2$"

```python
import sympy as sym
```

```python
# Indefinite integral
x = sym.symbols("x")

f = x
# this exclude the contants
sym.integrate(f)

p = sym.plotting.plot(f,show=False)
p.xlim = [0,1]
p.ylim = [0,1]
p.show()
```

```python
# Definite integral, for the value of "x", at positions 0, to 1
sym.integrate(f,(x,0,1))
```

```python
f = x**3 / (x-2)
int_f = sym.integrate(f)

p = sym.plotting.plot(f,show=False)
p.extend(sym.plotting.plot(int_f,(x,2.1,10),show=False,line_color='r'))

p[0].label = "$f(x) = %s$" %sym.latex(f)
p[1].label = "$\\int f(x) dx= %s$" %sym.latex(int_f)

p.ylim = [-200,200]
p.legend = True
p.show()
```
