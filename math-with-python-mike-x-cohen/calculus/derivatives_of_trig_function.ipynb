## Derivatives of trigonometry functions

$\frac{d}{dx} \cos(x) = -\sin(x)$

$\frac{d}{dx} -\sin(x) = -\cos(x)$

$\frac{d}{dx} -\cos(x) = \sin(x)$

$\frac{d}{dx} \sin(x) = \cos(x)$

```python
import sympy as sym
from IPython.display import display, Math
```

```python
x = sym.symbols("x")

print(sym.diff(sym.cos(x)))
print(sym.diff(-sym.sin(x)))
print(sym.diff(-sym.cos(x)))
print(sym.diff(sym.sin(x)))
```

```python
f = sym.cos(x)

# move till 2 cycles are complete
for i in range(0,8):
    display(Math("\\frac{d}{dx}%s = %s" %(sym.latex(f), sym.latex(sym.diff(f)))))
    f = sym.diff(f)
```

```python
import sympy.plotting.plot as symplot
```

```python
f = sym.cos(x)

for i in range(0,4):
    if i == 0:
        p = symplot(f,show=False,label=sym.latex(f))
    else: 
        p.extend(symplot(f,show=False,label=sym.latex(f)))
    f = sym.diff(f)

p.legend = True
p.xlim = [-3,3]
p.show()
```

## Exercise

```python
x,a = sym.symbols("x,a")
fax = sym.cos(x + sym.sin(x)) + a

# plotting the raw functions
for i in range(0,4): 
    if i == 0:
        p = symplot(fax.subs({a:i}),label="a="+str(i),show=False)
    else: 
        p.extend(symplot(fax.subs({a:i}),label="a="+str(i),show=False))

p.title = "The Functions"
p.legend = True
p.show()

# plotting the derivatives of functions, the derivatives of contant is zero, so all are graphs are overlapping each other
for i in range(0,4): 
    if i == 0:
        p = symplot(sym.diff(fax.subs({a:i})),label="a="+str(i),show=False)
    else: 
        p.extend(symplot(sym.diff(fax.subs({a:i})),label="a="+str(i),show=False))

p.title = "Their Derivatives"
p.legend = True
p.show()
```
