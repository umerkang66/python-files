## Partial Derivatives

If the function has more than one variables, then partial derivative is just the derivative of the function relative to one variable, by holding all the other variables constant

In the function that has more than one variables

$f(x,y) = 2xy^2$

Its partial derivative with respect to "x": here from the perspective of "x", "y" is a constant, so don't compute the "y", only compute (find derivative of) "x", because "y" can be anything

$\frac{\partial f}{\partial x} = f_x = 2y^2$

Its partial derivative with respect to "y": here from the perspective of "y", "x" is a constant, so don't compute the "x", only compute (find derivative of) "y", because "x" can be anything

$\frac{\partial f}{\partial y} = f_y = 4xy$

```python
import sympy as sym
from IPython.display import display, Math
```

```python
x,y = sym.symbols("x,y")
f = x**2 + x*y**2
display(Math("f = %s" %(sym.latex(f))))

# derivative with relative to "x"
print("Derivative with respect to \"x\"")
display(Math("\\frac{\\partial f}{\\partial x} = %s" %sym.latex(sym.diff(f,x))))

# derivative with relative to "y": if there is no "y" variable in term, so their whole derivative will be zero, with respect to "y"
print("Derivative with respect to \"y\"")
display(Math("\\frac{\\partial f}{\\partial x} = %s" %sym.latex(sym.diff(f,y))))
```
