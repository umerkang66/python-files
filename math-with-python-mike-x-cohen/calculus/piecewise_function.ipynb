```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
x = sym.symbols("x")

piece1 = 0
piece2 = -2*x
piece3 = x**3/10

# create piecewise functions according to given conditions
fx = sym.Piecewise((piece1,x<0), (piece2,(x>=0) &(x<3)), (piece3,x>=3))

fxx = sym.lambdify(x,fx)
x_range = np.linspace(-3,5,1234)

plt.plot(x_range, fxx(x_range))

plt.show()
```
