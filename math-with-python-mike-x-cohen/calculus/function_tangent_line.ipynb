```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
```

## Tangent Line

Tangent line that connect to the graph line of the function, just at one point

Formula to find the tangent line of a function in graph

$t = f'_a(x-x_a) + f_a$

```python
x = sym.symbols("x")

# define the function and its derivative
f = x**2
df = sym.diff(f)

# value at which to compute the tangent line
xa = 1

# get the function and derivative value x_a
fa = f.subs({x:xa})
dfa = df.subs({x:xa})

x_range = np.linspace(-2,2,200)
f_fun = sym.lambdify(x,f)(x_range)
df_fun = sym.lambdify(x,df)(x_range)

# compute teh tangent line
tanline = dfa * (x_range-xa) + fa

plt.plot(x_range,f_fun,label="f(x)")
plt.plot(x_range,tanline,label="Tangent")
plt.plot(xa,fa,'ro')

plt.axis("square")
plt.axis([-2,2,-.5,2])
plt.legend()

ax = plt.gca()
plt.plot(ax.get_xlim(),[0,0],'k--')
plt.plot([0,0],ax.get_ylim(),'k--')

plt.show()
```

```python
def compute_tangent(f,xa,bounds):
    # defining the derivative and values at point xa
    df = sym.diff(f)
    fa = f.subs(x,xa)
    dfa = df.subs(x,xa)

    # evaluate the tangent line
    x_range = np.linspace(bounds[0],bounds[1],200)
    return dfa * (x_range-xa) + fa
```
