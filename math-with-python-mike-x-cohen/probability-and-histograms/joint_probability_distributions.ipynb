```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import display
```

```python
n = 1000
x = np.random.randn(n)
y = np.random.randn(n)**2

plt.hist2d(x,y,30,vmax=10)

plt.xlabel("x")
plt.ylabel("$x^2$")
plt.show()
```

```python
sns.set()
sns.jointplot(x=x,y=y,kind="scatter").plot_joint(sns.kdeplot)

plt.xlabel("x")
plt.ylabel("$x^2$")
plt.show()
```
