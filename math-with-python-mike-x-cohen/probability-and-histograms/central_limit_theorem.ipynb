```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 2000

sig1 = np.sin(np.linspace(0,6*np.pi,n))
sig2 = np.random.rand(n)*2-1
sig3 = sig1 + sig2

# plot the individual signals and their distributions
fig,ax = plt.subplots(2,3,figsize=(10,4))

ax[0,0].plot(sig1)
ax[0,0].set_title("Sine wave")
ax[1,0].hist(sig1,100)

ax[0,1].plot(sig2,'.',alpha=.3)
ax[0,1].set_title("Uniform Numbers")
ax[1,1].hist(sig2,100)

ax[0,2].plot(sig3,'.',alpha=.3)
ax[0,2].set_title("Combined sig1,sig2")
ax[1,2].hist(sig3,100)

for i in np.matrix.flatten(ax):
    i.set_xticks([])
    i.set_yticks([])

plt.show()
```

```python
ns = np.arange(50,10000,100)
n_bins = 50
allhist = np.zeros((len(ns),n_bins))

for i in range(len(ns)):
    sig1 = np.sin(np.linspace(0,6*np.pi,ns[i]))
    sig2 = np.random.rand(ns[i])*2-1
    allhist[i,:],x = np.histogram(sig1+sig2,n_bins,density=True)


plt.pcolor(x[:-1],ns,allhist,shading="auto",vmax=.6,cmap="Purples")

plt.xlabel("Data Value")
plt.ylabel("Number of points in the sample")
plt.colorbar()
plt.show()
```
