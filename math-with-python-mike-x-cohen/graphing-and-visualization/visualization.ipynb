```python
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5

# b means blue, o means circle
plt.plot(x, y, "bo")

plt.axis("square")
# "x" axis will go from first number to second number
# "y" axis will go from third number to fourth number
plt.axis([-6, 6, -6, 6])
plt.grid()

plt.show()
```

```python
# for multiple coordinates
x = [-4, 5, 2, 4, 5, 6, 3]
y = [0, 3, 4, 5, 2, -3, 4]

for i in range(0, len(x)): 
    plt.plot(x[i], y[i], "o", label="point %s at [%s, %s]" %(i+1, x[i], y[i]))
    
plt.axis("square")
plt.axis([-8, 8, -8, 8])
plt.grid()
# enable it for labeling
plt.legend()
plt.show()
```

```python
plt.plot(4,2,"r+")

# gca is get current access, we can access the information about current graph
axis = plt.gca()
# "y" limit from both sides
y_lim = axis.get_ylim()
print(y_lim)
# lower bound of y_lim should remain the same, but change the upper bound to 5
axis.set_ylim([y_lim[0], 5])

plt.show()
```
