```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon
```

```python
y = np.array([ [1,1],[2,3],[3,1] ])
y1 = np.array([ [2,2],[2.5,4],[3.5,1] ])

# create polygon
p = Polygon(y, facecolor='g', alpha=.7)
p1 = Polygon(y1, facecolor='b', alpha=.7)

fig, ax = plt.subplots()
# add that polygon in "ax"
ax.add_patch(p)
ax.add_patch(p1)

ax.set_ylim([0, 4])
ax.set_xlim([0, 4])

plt.show()
```
