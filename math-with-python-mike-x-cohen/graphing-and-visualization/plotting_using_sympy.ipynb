```python
import sympy as sym
import numpy as np
import sympy.plotting.plot as symplot
```

```python
x = sym.symbols("x")
y = x**2

p = symplot(y, show=False)

p.xlim = [0, 50]
p[0].line_color = "m"
p.title = "This is a nice plot"

p.show()
```

```python
x,a = sym.symbols("x,a")

expr = a/x

n = 1
p = symplot(expr.subs(a, n), (x, -5, 5), show=False)# -5 is lower limit, 5 is upper limit
p[0].label = "$y = %s$" %sym.latex(expr.subs(a, n))

p2 = symplot(expr.subs(a, 2), (x, -5, 5), show=False)
p.extend(p2)
p[1].label = "$y = %s$" %sym.latex(expr.subs(a, 2))


p.ylim = [-5 , 5]
p.legend = True

p.show()
```
