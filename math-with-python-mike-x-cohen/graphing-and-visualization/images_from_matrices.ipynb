```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [ [1,2],[1,4] ]

plt.imshow(A)

plt.xticks([0, 1])
plt.yticks([0, 1])

plt.show()
```

```python
A = np.zeros((10, 14), dtype=int)

for i, row in enumerate(A):
    for j, col in enumerate(row): 
        A[i, j] = 3*i - 4*j

for i, row in enumerate(A):
    for j, col in enumerate(row): 
        # first is x-axis location, then y-axis location, in matplotlib the order is opposite, in matrix, first is row, then col, but in matplotlib, "x" axis is col, and "y" axis is rows, so first we have to deal with cols, then rows
        plt.text(j,i,A[i,j], horizontalalignment="center", verticalalignment="center")

plt.imshow(A)
plt.set_cmap("Purples")
plt.savefig("ThisIsANiceImage.png")
plt.show()
```

```python
N = 10
matrix = np.zeros((N, N), dtype=int)

for i in range(0, N): 
    for j in range(0, N): 
        matrix[i, j] = (-1)**(i + j)

plt.imshow(matrix)
plt.set_cmap("gray")
plt.tick_params(labelleft=False, labelbottom=False)
plt.show()
```
