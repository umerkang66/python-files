```python
import matplotlib.pyplot as plt
```

## Graphing Lines using StartEnd Notation

```python
p1 = [-3, -1]
p2 = [4, 4]

plt.plot([p1[0], p2[0]], [p1[1], p2[1]], color=[1, .2, 1], linewidth=5)

plt.axis("square")
plt.axis([-6, 6, -6, 6])
plt.show()
```

```python
x = 3
y = 5

plt.plot(x, y, "ro")
plt.plot([0, x], [0, y], "r")

plt.axis("square")
plt.axis([-6, 6, -6, 6])
plt.grid()

plt.plot([-6, 6], [0, 0], 'k')
plt.plot([0, 0], [-6, 6], 'k')

plt.show()
```

## Graphing lines using slope-intercept form

y = mx + b

where "x" is x-axis, "b" is the point of intersection at "y" and "m", "m" is how much you have to go up for each unit

```python
x = [-5, 5]
m = 2
b = 1

y = [0, 0]

for i, num in enumerate(x): 
    y[i] = m * num + b

plt.plot(x, y, label="y=%sx+%s" %(m, b))

plt.axis("square")
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
# get the horizontal x-axis limit
plt.plot(axis.get_xlim(), [0, 0], "k--")
plt.plot([0, 0], axis.get_ylim(), "k--")

plt.legend()

plt.show()
```

### Same Above Problem with Numpy array

```python
x = [-5, 5]
m = 2
b = 1

y = m * np.array(x) + b

print(x, y)
plt.plot(x, y, label="y=%sx+%s" %(m, b))

plt.axis("square")
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
# get the horizontal x-axis limit
plt.plot(axis.get_xlim(), [0, 0], "k--")
plt.plot([0, 0], axis.get_ylim(), "k--")

plt.legend()

plt.show()
```
