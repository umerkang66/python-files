```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(-3, 4)
y = np.zeros(len(x), dtype=int)

for i in range(0, len(x)): 
    y[i] = 2 - x[i]**2
    

plt.axis("square")
plt.axis([-8, 8, -8, 8])
plt.grid()

plt.plot([-8, 8], [0, 0], 'k')
plt.plot([0, 0], [-8, 8], 'k')


print(y)
plt.plot(x, y, 's-')
plt.show()
```

```python
x = np.linspace(-3, 3, 10)
y = 2 + np.sqrt(abs(x))

plt.plot(x, y, "ms-")
plt.show()
```

```python
powers = range(-1, 4)
x = np.linspace(-4, 4, 100)

for i in powers: 
    y = x**i
    plt.plot(x, y, label="$y = x^%s$" %i)

plt.legend()
plt.xlim(x[0], x[-1])
plt.ylim([-20, 20])
plt.show()
```
