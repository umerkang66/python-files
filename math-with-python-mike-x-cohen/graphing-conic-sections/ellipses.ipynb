```python
import numpy as np
import matplotlib.pyplot as plt
```

## Ellipses Formula

$\frac{(y-k)^2}{b^2} + \frac{(x-h)^2}{a^2} -1 = 0$

Here "a" and "b" are horizontal and vertical distances on graph, respectively. 
If "a" and "b" are equal, then formula becomes

$(y-k)^2 + (x-h)^2 - b^2 = 0$

which is exactly is the formula of circle

```python
# coordinates for the center of ellipse
h = 1
k = 2
# radius of ellipse
a = 2
b = 3

ax_lim = np.max((a,b)) + np.max((abs(h),abs(k)))
x = np.linspace(-ax_lim,ax_lim, 100)
X,Y = np.meshgrid(x,x)
# create the function
ellipse = ((X-h)**2 / a**2) + ((Y-k)**2 / b**2) - 1

plt.contour(X,Y,ellipse,0)
plt.plot(h,k,'go')

plt.gca().set_aspect("equal")
plt.show()
```
