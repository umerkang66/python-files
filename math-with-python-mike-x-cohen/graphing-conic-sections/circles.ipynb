```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plotting Circles in Python

Formula $\quad \Rightarrow \quad (y-k)^2 + (x-h)^2 = r^2$

where "r" is the radius, "h" is the vertical-axis, and "k" is the horizontal-axis, when circle centers at 0,0 coordinates, it means, h=0, and k=0

```python
h = 2
k = -3
r = 3

ax_lim = r + np.max((abs(h),abs(k)))
x = np.linspace(-ax_lim,ax_lim,100)
X,Y = np.meshgrid(x,x)

Fxy = (X-h)**2 + (Y-k)**2 - r**2

plt.contour(X,Y,Fxy,0)
plt.plot([-ax_lim, ax_lim], [0,0], 'k--')
plt.plot([0,0], [-ax_lim, ax_lim], 'k--')
plt.plot(h,k,'go')

# Both do same thing
# plt.axis("square")
plt.gca().set_aspect("equal")
plt.show()
```
