## Hyperbolas

Hyperbolas are just like ellipses, but they don't close in, but they start from starting point, toward their ends

Formula $\quad \Rightarrow \quad \frac{(y-k)^2}{b^2} - \frac{(x-h)^2}{a^2} -1 = 0$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = .5
h = 1
k = 2

ax_lim = 2* np.max((a,b)) + np.max((abs(h), abs(k)))
x = np.linspace(-ax_lim,ax_lim,100)
X,Y = np.meshgrid(x,x)

# create the function
hyperbola = ((X-h)**2 / a**2) - ((Y-k)**2 / b**2) - 1

plt.contour(X,Y,hyperbola,0)
plt.plot(h,k,'go')
plt.plot([-ax_lim,ax_lim],[0,0], '--', color=[.8,.8,.8])
plt.plot([0,0],[-ax_lim,ax_lim], '--', color=[.8,.8,.8])

plt.gca().set_aspect("equal")
plt.show()
```
