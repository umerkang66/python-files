```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# two matrices, X: 0,9  Y: 0,14
X,Y = np.meshgrid(range(0,10), range(0,15))

plt.subplot(121)
plt.pcolormesh(X,edgecolor="k")

plt.subplot(122)
plt.pcolormesh(Y,edgecolor="k")

plt.show()
```

```python
x = np.linspace(-np.pi,2*np.pi,40)
y = np.linspace(0,4*np.pi,72)

X,Y = np.meshgrid(x,y)

Fxy = np.cos(X) + np.sin(Y)

plt.imshow(Fxy)
plt.show()
```
