### Test whether the outer product is matrix commutative, and scaler multiplication commutative

Outer product is not Matrix commutative, but it is Scaler-multiplication commutative

### Matrix commutative 
np.outer(v,w) == np.outer(w,v)

### Scaler-multiplication commutative
$cvw^T = vcw^T = vw^Tc$

```python
# Matrix commutative: not matrix commutative
v = np.arange(1,11)
w = np.arange(1,5)

print(np.outer(v,w))
print(np.outer(w,v))
```

```python
# Scaler Multiplication
s = 4

res1 = s*np.outer(v,w)
res2 = np.outer(s*v,w)
res3 = np.outer(v,s*w)
res4 = np.outer(v,w)*s

print(res3-res4)
```
