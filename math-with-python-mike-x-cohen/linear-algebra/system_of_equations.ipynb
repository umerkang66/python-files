## System of equations

$2x + y - z = 1$

$3x + 4y + 2z = 13$

$x - 5y - 2z = 0$

The above equations become: if we implement the dot product of the left side of equation, we can get the original equations

$$\begin{bmatrix} 
	2 & 1 & -1  \\
	3 & 4 &  2  \\
	1 & -5 & -2 \\
\end{bmatrix} \begin{bmatrix} 
	x \\
	y \\
	z \\
\end{bmatrix} = \begin{bmatrix} 
	1  \\
	13 \\
	0 \\
\end{bmatrix} $$

Can also be written as:

$$X \times \beta = y$$

To solve for "$\beta$", we can use inverse method, or use "Least-square solution", and its formula is

$$\beta = (X^TX)^{-1}X^Ty$$

where $X^T$ is transpose version of $X$

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
X = np.array([ [2,1,-1],[3,4,2],[1,-5,-2] ])
y = np.array([ 1,13,0 ])

beta1 = np.linalg.inv(X.T@X) @ X.T @ y
# this will solve for Beta
beta2 = np.linalg.solve(X,y)

print(beta1,beta2)
```

```python
# system 1
X1 = np.array([ [3,-1],[-1,1] ])
y1 = np.array([ 6,2 ])

# system 2
X2 = np.array([ [3,-1],[1.5,-.5] ])
y2 = np.array([ 6,3 ])

# system 3
X3 = np.array([ [3,-1],[1.5,-.5] ])
y3 = np.array([ 6,2 ])
```

```python
xlim = np.array([0,10])
# system 1
# convert to => y = mx+b
yy1 = -X1[0,0]/X1[0,1]*xlim + y1[0]/X1[0,1]
yy2 = -X1[1,0]/X1[1,1]*xlim + y1[1]/X1[1,1]

# plot
plt.plot(xlim,yy1,'b',label="eq 1")
plt.plot(xlim,yy2,'r',label="eq 2")

# solve
b = np.linalg.solve(X1,y1)
plt.plot(b[0],b[1],'ko',markersize=10,label="solution")

plt.legend()
plt.grid()
plt.show()
```
