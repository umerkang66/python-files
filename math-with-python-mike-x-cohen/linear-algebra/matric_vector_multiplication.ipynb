```python
import numpy as np
import matplotlib.pyplot as plt
```

When we multiply vector with matrix, it will expand or shrink the vector, and it will change the direction in plot

Matrix with vector Multiplication. It is actually a dot product

$$\begin{bmatrix} 
	.5 & 1 \\
	1 & .5\\
\end{bmatrix} \begin{bmatrix} 
	1 \\
	2 \\
\end{bmatrix} = \begin{bmatrix} 
	2.5 \\
	2 \\
\end{bmatrix} $$

```python
M = np.array([
    [.5,1],
    [1,.5]
])
v = np.array([1,2])

Mv = M@v
print(Mv)

plt.plot([0,v[0]], [0,v[1]], linewidth=2, label="v")
plt.plot([0,Mv[0]], [0,Mv[1]], 'r--', linewidth=2, label="Mv")

plt.axis("square")
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()
```

```python
v_comp = np.linspace(-2,2,40)
M = np.array([
    [.5,1],
    [1,.5]
])

for a in v_comp:
    v = np.array([1,a])
    Mv = M@v
    # plotting
    plt.plot([0,v[0]], [0,v[1]])
    plt.plot([0,Mv[0]], [0,Mv[1]])

plt.axis("square")
plt.show()
```
