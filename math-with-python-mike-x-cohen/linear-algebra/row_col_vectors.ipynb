```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
```

```python
row_vec = np.array([4,2])
col_vec = np.array([ [-2],[3] ])

display(Math(sym.latex(sym.sympify(row_vec))))
display(Math(sym.latex(sym.sympify(col_vec))))
```

```python
# plotting the vectors
# for algebraic operations, row vectors and not the same thing as the column vectors, but in plot, they are drawn the same
plt.plot([0,row_vec[0]], [0,row_vec[1]], 'r', label="Row Vector")
plt.plot([0,col_vec[0,0]], [0,col_vec[1,0]], 'b', label="Column Vector")

plt.axis("square")
plt.axis([-5,5,-5,5])
plt.legend()
plt.grid()

plt.show()
```

## Adding and Subtracting Vectors

```python
v1 = np.array([-1,2])
v2 = np.array([1,1])

v3a = v1 + v2
v3b = np.add(v1,v2)

v3c = np.zeros(2)
for i in range(0,2):
    v3c[i] = v1[i] + v2[i]

print(v3a,v3b,v3c)
```

### Geometrically addition of vectors

```python
v1 = np.array([-1,2])
v2 = np.array([1,1])
v3 = v1 + v2

plt.plot([0,v1[0]],[0,v1[1]], linewidth=3, label="$v_1$")
plt.plot([v1[0],v2[0]+v1[0]] ,[v1[1],v2[1]+v1[1]], linewidth=3, label="$v_2$")

plt.plot([0,v3[0]] ,[0,v3[1]], linewidth=3, label="$v_3$")

s1 = sym.latex(sym.sympify(v1))
s2 = sym.latex(sym.sympify(v2))
s3 = sym.latex(sym.sympify(v3))
display(Math("%s + %s = %s" %(s1,s2,s3)))

plt.legend()
plt.axis("square")
plt.axis([-2,2,-2,4])

plt.grid()
plt.show()
```

## Vector Scaler Multiplication

```python
# this can interpreted x,y coordinates in graph
v = np.array([.5,1])
scaler = [1,-.5,2,.5]

# the scaled version of vector, will never change, direction as you can see in graph (it can still goes negative)
for si in scaler:
    scaled_v = si*v
    plt.plot([0,scaled_v[0]], [0,scaled_v[1]],'o-',linewidth=3,label="$\\lambda = %g$" %si)

plt.legend()
plt.axis("square")
plt.axis([-3,3,-3,3])
plt.grid()

plt.show()
```
