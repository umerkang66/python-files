```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ein = Image.open('einstein.jpg')
ein = np.mean(ein,2)
print(np.shape(ein))

plt.imshow(ein)
plt.show()
```

```python
U,s,V = np.linalg.svd(ein)

print(np.shape(U))
print(np.shape(s))
print(np.shape(V))

fig,ax = plt.subplots(1,3)

ax[0].imshow(U)
ax[0].set_title("U")

ax[1].imshow(np.diag(np.log(s)))
ax[1].set_title("$\\Sigma$")

ax[2].imshow(V)
ax[2].set_title("V")

plt.show()
```

```python
plt.plot(s,'ks-',markerfacecolor='w')

plt.title("Eigen spectrum of einstein")
plt.xlim([-1,50])
plt.show()
```

## Exercise

```python
S = np.zeros(np.shape(ein))
for i in range(0,len(s)):
    S[i,i] = s[i]

reconstructed_ein = U@S@V

plt.subplot(1,3,1)
plt.imshow(ein)
plt.title("Original")

plt.subplot(1,3,2)
plt.imshow(reconstructed_ein)
plt.title("SVD Version")

# shuffling the values
randorder = np.random.permutation(len(s))
for i in range(0,len(s)):
    S[i,i] = s[randorder[i]]

reconstructed_ein = U@S@V

plt.subplot(1,3,3)
plt.imshow(reconstructed_ein)
plt.title("Shuffled values")

plt.show()
```
